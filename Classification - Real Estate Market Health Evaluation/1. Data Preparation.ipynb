```python
import pandas as pd

df = pd.read_csv("MarketHealthIndex_City.csv", engine="python")
df.info()
```

```python
df = df.drop(['PrevForeclosed','ForeclosureRatio','StockOfREOs'],axis=1)
```

```python
df1 = pd.read_csv("City_Zri_AllHomesPlusMultifamily_Summary.csv", engine="python")
df1.info()
```

```python
df1 = df1.drop(['Date', 'State', 'Metro', 'County', 'SizeRank', 'MoM', 'YoY', 'ZriRecordCnt'],axis=1)
```

```python
# merge 
df = df.merge(df1, on="RegionName", how="inner", left_index=True, right_index=True)
df.info()
df.head()
```

```python
df2 = pd.read_csv("BreakEven_2017-03.csv", engine="python")
df2 = df2[(df2.RegionType!='Neighborhood')]
df2.info()
```

```python
df2 = df2.drop(['RegionType', 'RegionID', 'SizeRank', 'City', 'CountyName', 'Metro', 'CBSATitle', 'StateName', 'Breakeven'],axis=1)
```

```python
df = df.merge(df2, on="RegionName", how="inner", left_index=True, right_index=True)
df.info()
df.head()
```

```python
df = df.reset_index()
df.drop('index', axis=1, inplace=True)
df.info()
```

```python
df3 = pd.read_csv("NESummary_2017Q1_Public.csv", engine="python")
df3 = df3[(df3.RegionType=='City')]
df3.info()
```

```python
df3 = df3.drop(['RegionType', 'Negative Equity Percent of Homes w/Mortgage', 'Number of Homes at Least 90 Days Late', 'Effective Negative Equity Rate'],axis=1)
```

```python
df3=df3.reset_index()
df3.drop('index', axis=1, inplace=True)
df3.info()
```

```python
df = df.merge(df3, how="left", on="RegionName", left_index=True, right_index=True)
df.info()
df.head()
```

```python
df.to_csv('DataMerged.csv', header=True, index=True)
```
